```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
emotions = ['admiration', 'amusement', 'anger', 'annoyance', 'approval', 'caring', 'confusion', 'curiosity', 'desire', 'disappointment', 'disapproval',
            'disgust', 'embarassment', 'excitement', 'fear', 'gratitude', 'grief', 'joy', 'love', 'nervousness', 'optimism', 'pride', 'realization', 'relief',
            'remorse', 'sadness', 'surprise', 'neutral']

results = pd.read_csv('SportSentiment_merged_output_2nd.csv')
#results2 = pd.read_csv('SportSentiment_merged_output_2nd.csv')

for ind in results.index:
  results['predicted_emotion'][ind] = emotions[results['predicted_emotion'][ind]]
  results['predicted_emotion_2nd'][ind] = emotions[results['predicted_emotion_2nd'][ind]]


relevant = results[['sport', 'phase', 'category', 'predicted_emotion', 'predicted_emotion_2nd']]
relevant.head()
```

```python
#option to replace primary 'neutral' emotion with secondary emotion

#for ind in relevant.index:
  #if relevant['predicted_emotion'][ind] == 'neutral':
   # relevant['predicted_emotion'][ind] = relevant['predicted_emotion_2nd'][ind]
```

```python
#plot of fans of favored team

favored_1 = relevant[relevant['category'].isin(['1F'])][['category', 'predicted_emotion']]
favored_2 = relevant[relevant['category'].isin(['2F'])][['category', 'predicted_emotion']]
favored_3 = relevant[relevant['category'].isin(['3F'])][['category', 'predicted_emotion']]
favored_4 = relevant[relevant['category'].isin(['4F'])][['category', 'predicted_emotion']]

favored_1 = favored_1.reset_index(drop=True)
favored_2 = favored_2.reset_index(drop=True)
favored_3 = favored_3.reset_index(drop=True)
favored_4 = favored_4.reset_index(drop=True)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(18)
fig.set_figwidth(19)
fig.suptitle("Predicted Emotions of Favored Team Supporters", x=0.5, y=0.92)

favored_1["predicted_emotion"].value_counts(True).plot(ax = ax1, kind="bar", title="1F")
favored_2["predicted_emotion"].value_counts(True).plot(ax = ax2, kind="bar", title="2F")
favored_3["predicted_emotion"].value_counts(True).plot(ax = ax3, kind="bar", title="3F")
favored_4["predicted_emotion"].value_counts(True).plot(ax = ax4, kind="bar", title="4F")

fig.savefig("F_NoNeutral.jpg")
```
